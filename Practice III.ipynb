```python
# Importing necessary libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
from sklearn.datasets import load_iris
```

```python
# Loading the Iris dataset
data = load_iris()
X = data.data
y = data.target
feature_names = data.feature_names
```

```python
# Standardizing the data (mean=0, std=1)
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
# Applying PCA to reduce the data to 2 dimensions
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)
```

```python
# Creating a DataFrame to view the results
df = pd.DataFrame(data=X_pca, columns=['PC1', 'PC2'])
df['Target'] = y
```

```python
# Plotting the PCA results
plt.figure(figsize=(8,6))
plt.scatter(df['PC1'], df['PC2'], c=df['Target'], cmap='viridis', edgecolor='k', s=100)
plt.title('PCA of Iris Dataset')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.colorbar(label='Target')
plt.show()
```

```python
# Explained variance ratio (the importance of each principal component)
print('Explained Variance Ratio:', pca.explained_variance_ratio_)
```
