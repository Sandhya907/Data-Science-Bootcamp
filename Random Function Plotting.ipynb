```python
# Import necessary libraries
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Enable inline plotting
%matplotlib inline  
```

```python
# Generate x values
x = np.linspace(-10, 10, 400)
```

```python
# Define a function
y = np.sin(x) + np.cos(0.5 * x) + 0.2 * x
```

```python
# Plot the function
plt.figure(figsize=(8, 5))
plt.plot(x, y, label="y = sin(x) + cos(0.5x) + 0.2x", color='b', linewidth=2)
```

```python
# Customize the plot
plt.axhline(0, color='black', linewidth=0.5, linestyle="--")
plt.axvline(0, color='black', linewidth=0.5, linestyle="--")
plt.grid(True, linestyle="--", alpha=0.6)
plt.legend()
plt.xlabel("X values")
plt.ylabel("Y values")
plt.title("Random Function Plot")

# Display the plot
plt.show()
```
