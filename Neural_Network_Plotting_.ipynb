```python
import matplotlib.pyplot as plt
import networkx as nx
```

```python
def plot_nn():
    Gr = nx.DiGraph()

# Define nodes
    input_nodes = ['i1', 'i2', 'i3', 'i4']
    hidden_nodes = ['h1', 'h2', 'h3']
    output_nodes = ['o']

# Add nodes to the graph
    Gr.add_nodes_from(input_nodes, layer='input')
    Gr.add_nodes_from(hidden_nodes, layer='hidden')
    Gr.add_nodes_from(output_nodes, layer='output')

# Define edges between layers
    for node_in in input_nodes:
        for node_hidden in hidden_nodes:
            Gr.add_edge(node_in, node_hidden)  # Connect inputs to hidden layer

    for node_hidden in hidden_nodes:
        Gr.add_edge(node_hidden, output_nodes[0])  # Connect hidden to output layer

# Assign colors based on layer type
    node_colors = ['lightblue' if Gr.nodes[node].get('layer') == 'input' else
                   'lightgreen' if Gr.nodes[node].get('layer') == 'hidden' else
                   'orange' for node in Gr.nodes]

    return Gr, node_colors
```

```python
# Call function to generate the graph
Gr, node_colors = plot_nn()
```

```python
# Define fixed positions for nodes
pos = {
    'i1': (0, 0), 'i2': (0, 1), 'i3': (0, 2), 'i4': (0, 3),
    'h1': (1, 0.5), 'h2': (1, 2), 'h3': (1, 3),
    'o': (2, 2)
}
```

```python
# Draw the neural network graph
plt.figure(figsize=(10, 8))
nx.draw(Gr, pos, with_labels=True, node_size=3500,
        node_color=node_colors, font_size=12,
        font_weight='bold', edge_color='black')
plt.title("Simple Neural Network Graph Architecture")
plt.show()
```
